## Cost of Employee Attrition

```python
#assumes cost to replace employee is 75% of annual salary
#df['Cost_to_replace'] = df.MonthlyIncome*12*.75

cost = []
idx  = 0
for i in df['JobLevel']:
    if i == 1:
        cost.append(df.MonthlyIncome[idx]*12*.5)
    elif i == 2:
        cost.append(df.MonthlyIncome[idx]*12*.75)
    elif i == 3:
        cost.append(df.MonthlyIncome[idx]*12*1)
    elif i == 4:
        cost.append(df.MonthlyIncome[idx]*12*1.5)
    elif i == 5:
        cost.append(df.MonthlyIncome[idx]*12*2)
    idx += 1
        
df['Cost_to_replace'] = cost
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from imblearn.over_sampling import SMOTE
```

```python
df.OverTime = df.OverTime.astype('category')
df.StockOptionLevel = df.StockOptionLevel.astype('category')
df.JobLevel = df.JobLevel.astype('int')
```

```python
# Labels are the values we want to predict
labels = np.array(df['Attrition_Bio'])

# Remove the labels from the features
# axis 1 refers to the columns
features = df[['Age', 'BusinessTravel', 'DailyRate', 'Department',
       'DistanceFromHome', 'Education', 'EducationField',
       'EnvironmentSatisfaction', 'Gender', 'HourlyRate', 'JobInvolvement',
       'JobLevel', 'JobRole', 'JobSatisfaction', 'MaritalStatus',
       'MonthlyIncome^2', 'MonthlyRate', 'NumCompaniesWorked',
       'OverTime', 'PercentSalaryHike', 'PerformanceRating',
       'RelationshipSatisfaction', 'StockOptionLevel', 'TotalWorkingYears',
       'TrainingTimesLastYear', 'WorkLifeBalance', 'YearsAtCompany']]

#df[[ 'OverTime', 'DailyRate','JobRole','JobInvolvement',
              #'WorkLifeBalance', 'StockOptionLevel', 'MaritalStatus', 
                     #'Age','NumCompaniesWorked','DistanceFromHome','EnvironmentSatisfaction']]




#df[['YearsAtCompany','MaritalStatus','JobRole','StockOptionLevel','JobLevel', 'JobSatisfaction',
       #'EnvironmentSatisfaction', 'OverTime', 'MonthlyIncome', 'Age', 'Department']]

features = pd.get_dummies(features)

# Saving feature names for later use
feature_list = list(features.columns)

# Convert to numpy array
features = np.array(features)

# Using Skicit-learn to split data into training and testing sets
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.2, random_state = 42)

#from imblearn.over_sampling import RandomOverSampler
#ros = RandomOverSampler(random_state=0)
#smote_train, smote_target = ros.fit_resample(train_features,train_labels)

oversampler=SMOTE(sampling_strategy='auto',k_neighbors = 5, random_state=0)
smote_train, smote_target = oversampler.fit_sample(train_features,train_labels)
```

```python
from sklearn.cluster import KMeans

#Let's select the optimum number of clusters for k-means classification
from sklearn.cluster import KMeans
#Within-Cluster-Sum-of-Squares WCSS
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++',
                    max_iter = 400, n_init = 10, random_state = 0)
    kmeans.fit(x)
    print(kmeans.inertia_)
    wcss.append(kmeans.inertia_)
    
#Plotting the results onto a line graph to observe 'The elbow'
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Association')
plt.ylabel('WCSS') #within cluster sum of squares
plt.show()
```

```python
def cross_validate(model, features, labels, cv = 10, scoring='accuracy', n_jobs=-1):
    from sklearn.model_selection import cross_val_score
    #Cross Validation
    Accuracy_scores = cross_val_score(model, features, labels, cv=cv, 
                                  scoring=scoring, 
                                  n_jobs=n_jobs) 

    # Compute the K-folds CV RMSE
    print(Accuracy_scores)
    Accuracy_scores_CV = (Accuracy_scores.mean())
    Accuracy_scores_SD = (Accuracy_scores.std())
    print('CV Accuracy: {:.2f}'.format(Accuracy_scores_CV))
    print('CV Accuracy SD: {:.2f}'.format(Accuracy_scores_SD))
```
