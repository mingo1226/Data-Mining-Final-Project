## 2.1 Unsupervised Learning - Clustering

```python
#using KMeans to cluster
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
#throwing away some useless vars in our clustering
```

```python
#construct an Elbow method

def elbow_method(data):
    #standardize data
    std_tool = StandardScaler()
    x_std = std_tool.fit_transform(data)
    #test WCSS with each # of clusters (15clusters MAX)
    wcss = []
    for i in range(1, 31):
        kmeans = KMeans(n_clusters = i, init = 'k-means++',
                    max_iter = 400, n_init = 10, random_state = 0)
        kmeans.fit(x_std)
        wcss.append(kmeans.inertia_)
    return wcss
```

```python
#design a pipeline for PCA.
def PCA_pipeline(data_input):
    from sklearn.decomposition import PCA
    from sklearn.preprocessing import StandardScaler
    data = data_input
    x = data
    #standardize the data
    std_scaler = StandardScaler()
    x_std = std_scaler.fit_transform(x)
    #fit a pca
    pca_fit = PCA(0.85)
    x_transformed = pca_fit.fit_transform(x_std)
    return x_transformed
```

### 2.2.1 Logistic Regression

```python
temp['Cost_to_replace'] = df.MonthlyIncome*12*.75
```

#### Logistic Regression: 
1. NA & oversampling & variable encoding; feature engineering;
2. L1 & L2 for multi-collinearity; 
3. get probability
4. run K-Means for flat Elbow
5. get summary results for each group -> SUM(probability * attrition cost)

```python
#design a pipeline for LR.
def LR_pipeline(data_input):
    from imblearn.over_sampling import SMOTE
    from sklearn.preprocessing import StandardScaler
    from sklearn.model_selection import train_test_split
    from sklearn.linear_model import LogisticRegression
    data = data_input
    attrition_cost = data['Cost_to_replace']
    #drop it for sake of perfect multico with monthly income
    data = data.drop(['Cost_to_replace'],axis = 1)
    #train-test split
    x = data[data.columns[~data.columns.isin(['Attrition'])]]
    y = data['Attrition']
    #standardize the data
    std_scaler = StandardScaler()
    x_std = std_scaler.fit_transform(x)
    #over-sampling using SMOTE
    smote_tool = SMOTE(random_state = 0)
    x_resampled, y_resampled = smote_tool.fit_resample(x_std, y)
    #train_test split
    x_train, x_test, y_train, y_test = train_test_split(x_resampled, y_resampled,random_state = 0, test_size = 0.2)
    #construct LR model
    lr_model = LogisticRegression(penalty = 'l2', solver = 'lbfgs')
    model = lr_model.fit(x_train,y_train)
    return model
```

```python
#design a pipeline for RF.
def RF_pipeline(data_input):
    from imblearn.over_sampling import SMOTE
    from sklearn.preprocessing import StandardScaler
    from sklearn.model_selection import train_test_split
    from sklearn.ensemble import RandomForestClassifier
    data = data_input
    attrition_cost = data['Cost_to_replace']
    #drop it for sake of perfect multico with monthly income
    data = data.drop(['Cost_to_replace'],axis = 1)
    #train-test split
    x = data[data.columns[~data.columns.isin(['Attrition'])]]
    y = data['Attrition']
    #standardize the data
    std_scaler = StandardScaler()
    x_std = std_scaler.fit_transform(x)
    #over-sampling using SMOTE
    smote_tool = SMOTE(random_state = 0)
    x_resampled, y_resampled = smote_tool.fit_resample(x_std, y)
    #train_test split
    x_train, x_test, y_train, y_test = train_test_split(x_resampled, y_resampled,random_state = 0, test_size = 0.2)
    #construct LR model
    rf = RandomForestClassifier(n_estimators = 1000, max_depth = 5, min_samples_leaf = 2, max_features = .5,
                            oob_score=True, random_state = 0)
    model = rf.fit(x_train,y_train)
    return model
```
